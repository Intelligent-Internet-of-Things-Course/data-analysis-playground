# CSV Exercise - 1

- Read smarthome_power_consumption_small.csv
- Skip the first row and store it in a different list called headers
- Store all the other rows in a matrix

```python
#Import and open the file
import csv
import os

data_folder = 'data' 
data_file = 'smarthome_power_consumption_small.csv'

file_path = os.path.join(data_folder,data_file)

data_matrix = []
header = []
with open(file_path,"r") as f :
    first_line = f.readline()
    header = first_line.replace("\n","").split(",")
    #print(header.index("Age"))
    for row in f.readlines():
        row_fields = row.replace("\n","").split(",")
        data_matrix.append(row_fields)

print("\nHeader: \n")
print(header)
        
print("\nData Matrix: \n")
print(data_matrix)
```

```python
#Import and open the file
import csv
import os

data_folder = 'data' 
data_file = 'smarthome_power_consumption_small.csv'

file_path = os.path.join(data_folder,data_file)

data_matrix = []
header = []
with open(file_path,"r") as f :
    first_line = f.readline()
    header = first_line.replace("\n","").split(",")

header_string = " " 
request = input("Specify a target column among " + header_string.join(header) + " ---> ")
target_column = request.split(",")[0]
target_column_index = header.index(target_column)

print("\nLoading data for column: " + target_column + " Index: " + str(target_column_index))

with open(file_path,"r") as f :
    #Skip first row
    first_line = f.readline()
    for row in f.readlines():
        row_fields = row.replace("\n","").split(",")
        data_matrix.append(row_fields[target_column_index])
            
print(data_matrix)
```

# CSV Exercise - 3

- Read only the first row of smarthome_power_consumption_small.csv
- Ask to the user the target column name of interest among the availables
- Read data row together with the target column from the specified data source file
- Write into a new CSV file target column and the date

```python
#Import and open the file
import csv
import os

data_folder = 'data' 
data_file = 'smarthome_power_consumption_small.csv'

file_path = os.path.join(data_folder,data_file)

data_matrix = []
header = []
with open(file_path,"r") as f :
    first_line = f.readline()
    header = first_line.replace("\n","").split(",")

header_string = " " 
request = input("Specify a target column among " + header_string.join(header) + " ---> ")
target_column = request.split(",")[0]
target_column_index = header.index(target_column)

print("\nLoading data for column: " + target_column + " Index: " + str(target_column_index))

with open(file_path,"r") as f :
    #Skip first row
    first_line = f.readline()
    for row in f.readlines():
        row_fields = row.replace("\n","").split(",")
        data_matrix.append([row_fields[0], row_fields[target_column_index]])

#Define output file header
target_header = [header[0], target_column]

#Open the file in write mode
with open('data/output_voltage.csv', 'w', newline='') as f:
    #Create the CSV writer
    writer = csv.writer(f)
    #Write CSV Header
    writer.writerow(target_header)
    #Iterate in order to populate the CSV file through available data rows
    for data_row in data_matrix:
        writer.writerow(data_row)
        
print("CSV File Correctly Created !")
```
